# Ames Iowa Housing Project

## Overview
The project scope was based around a city in the state of Iowa. The city is called Ames. We are using the given information on houses or properties sold between 2006 and 2010. The goal of the project was to use all current skills to refine the data set into a workable data frame to be used for modeling and gathering insights into the housing market in Ames during the given period of time.

## Conclusions
After parsing through the refined data frame, I chose to target land developers. In particular I was looking at housing, and that included Multi-Family and Simgle Family dwellings. 

While there are many different paths that can be taken, I felt that going the path of Multi-Family housing was the best choice based on the size of the city, the work force, and that the town itself holds a college campus. With on average large lot space, it is possible to provide many dwellings in a tighter space. Allowing the town to keep many of its large open space feel to most of the city. While also giving it the necessary spaces for population and economic growth. Plus with cities that have a small town feel to them. The last thing you want to do is overrun them with cookie cutter housing neighborhoods.
